```python
import json
import os
from pathlib import Path

import ee
import ee.deserializer
import geemap

from afolu.assets.constants import LABEL_LIST
```

```python
zone = "small"
```

```python
ee.Initialize(project="ee-ursa-test")
```

```python
data_path = Path(os.getenv("DATA_PATH"))
out_path = data_path / "generated"
```

```python
with open(out_path / zone / "bbox" / "ee.json") as f:
    bbox = ee.deserializer.decode(json.load(f))
```

```python
bbox_rasterized = ee.image.Image.constant(0).paint(bbox, 1).selfMask()
bbox_area = (
    ee.image.Image.pixelArea()
    .multiply(bbox_rasterized)
    .reduceRegion(ee.reducer.Reducer.sum(), geometry=bbox, scale=30, maxPixels=1e10)
    .getInfo()["area"]
)
bbox_area
```

```python
for start_year, end_year in zip(range(2000, 2022), range(2001, 2023), strict=False):
    s = 0
    for start_label in LABEL_LIST:
        for end_label in LABEL_LIST:
            with open(
                out_path
                / "amazon"
                / "transition"
                / "table_split"
                / f"{start_label}-{end_label}"
                / f"{start_year}_{end_year}.txt",
                encoding="utf8",
            ) as f:
                s += float(f.read())

    print(s / bbox_area["area"])
```

```python
for year in range(2000, 2023):
    s = 0
    for label in LABEL_LIST:
        with open(
            out_path / "amazon" / "area" / "area_table" / f"{label}" / f"{year}.txt",
            encoding="utf8",
        ) as f:
            s += float(f.read())

    print(s / bbox_area["area"])
```
