```python
import ee
import ee.deserializer
import geemap
import json
import os

from pathlib import Path
```

```python
ee.Initialize(project="ee-ursa-test")
```

```python
out_path = Path(os.environ["DATA_PATH"]) / "generated"
```

```python
with open(out_path / "small" / "glc30.json") as f:
    img = ee.deserializer.decode(json.load(f))

with open(out_path / "small" / "bbox" / "ee.json") as f:
    bbox = ee.deserializer.decode(json.load(f))
```

```python
geemap.download_ee_image(img, "./glc30.tif", region=bbox, scale=30, crs="EPSG:4326")
```

```python
m = geemap.Map()
m.addLayer(img)
m
```
