```python
import os

import geopandas as gpd

from pathlib import Path
```

```python
data_path = Path(os.environ["DATA_PATH"])
```

```python
df = gpd.read_file(data_path / "initial" / "boundaries" / "GUY+Georgetown").to_crs("ESRI:102033")
union = gpd.GeoDataFrame(geometry=[df["geometry"].buffer(5).union_all()], crs="ESRI:102033")
```

```python
union.to_file("./temp.gpkg")
```
