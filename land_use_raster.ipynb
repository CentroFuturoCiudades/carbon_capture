```python
import ee
import ee.deserializer
import json
import geemap
import os

from dotenv import load_dotenv
from pathlib import Path
```

```python
load_dotenv()
```

```python
ee.Initialize(project="ee-ursa-test")
```

```python
data_path = Path(os.getenv("DATA_PATH"))
out_path = data_path / "generated"
```

```python
ZONE = "BRA+São Luís"
```

```python
with open(out_path / "small" / "area" / "raster" / "2020" / f"{ZONE}.json", "r") as f:
    img = ee.deserializer.decode(json.load(f))

with open(out_path / "small" / "bbox" / "ee" / f"{ZONE}.json") as f:
    bbox = ee.deserializer.decode(json.load(f))

geemap.download_ee_image(img, "./area.tif", region=bbox, scale=30, crs="EPSG:4326")
```

```python
img.bandNames().getInfo()
```

```python
m = geemap.Map()
m.addLayer(img, {}, "img")
m
```
