```python
import os
import shutil
import tempfile
from pathlib import Path
```

```python
ZONE = "PER+Coronel Portillo"
```

```python
data_path = Path(os.environ["DATA_PATH"])
out_path = data_path / "generated"
```

```python
with tempfile.TemporaryDirectory() as tmpdir:
    tmp_path = Path(tmpdir)
    (tmp_path / "raster").mkdir()
    (tmp_path / "forecast" / "area").mkdir(parents=True)
    (tmp_path / "forecast" / "emissions").mkdir(parents=True)

    shutil.copy(
        out_path / "small" / "area" / "table_merged" / f"{ZONE}.csv",
        tmp_path / "area.csv",
    )
    shutil.copy(
        out_path / "small" / "transition" / "cube" / f"{ZONE}.csv",
        tmp_path / "cube.csv",
    )
    shutil.copy(
        out_path / "small" / "emissions" / f"{ZONE}.csv",
        tmp_path / "emissions.csv",
    )
    shutil.copy(
        out_path
        / "small"
        / "transition"
        / "table_fixed_extra"
        / "2000_2020"
        / f"{ZONE}.csv",
        tmp_path / "table_extra.csv",
    )
    shutil.copy(
        out_path / "small" / "transitions_emissions" / f"{ZONE}.csv",
        tmp_path / "transitions_emissions.csv",
    )
    shutil.copy(
        out_path / "small" / "stats" / "population" / f"{ZONE}.csv",
        tmp_path / "population.csv",
    )
    shutil.copy(
        out_path / "small" / "stats" / "built_area" / f"{ZONE}.csv",
        tmp_path / "built_area.csv",
    )
    shutil.copy(
        out_path / "small" / "raster" / "built_area" / f"{ZONE}.tif",
        tmp_path / "raster" / "built_area.tif",
    )
    shutil.copy(
        out_path / "small" / "raster" / "population" / f"{ZONE}.tif",
        tmp_path / "raster" / "population.tif",
    )
    shutil.copy(
        out_path / "small" / "raster" / "built_binary" / f"{ZONE}.tif",
        tmp_path / "raster" / "built_binary.tif",
    )
    shutil.copy(
        out_path / "small" / "bbox" / "shapely" / f"{ZONE}.gpkg",
        tmp_path / "bbox.gpkg",
    )

    for prefix in ["normal", "fast", "slow"]:
        shutil.copy(
            out_path / "small" / "forecast" / "area" / f"{prefix}" / f"{ZONE}.csv",
            tmp_path / "forecast" / "area" / f"{prefix}.csv",
        )
        shutil.copy(
            out_path / "small" / "forecast" / "emissions" / f"{prefix}" / f"{ZONE}.csv",
            tmp_path / "forecast" / "emissions" / f"{prefix}.csv",
        )

    shutil.make_archive("./compressed", "zip", tmp_path)
```
